```python
import pandas as pd
import re
file_path = '/kaggle/input/lichess-dataset-1/lichess_db_standard_rated_2013-01.csv'

df = pd.read_csv(file_path)
def remove_urls(text):
    url_pattern = r'https?://\S+'
    return re.sub(url_pattern, '', text).strip()

df["Event"]=df['Event'].apply(remove_urls)

grouped = df.groupby("Event")

for name, group in grouped:
    file_name = f"{name.replace(' ', '_').replace('/', '_')}.csv"
    individual_game_df=group
    df_unique=individual_game_df.drop_duplicates(subset=['Game Number'], keep='first')
    group.to_csv(file_name, index=False)
    print(f"Saved '{name}' to '{file_name}' with {df_unique.shape[0]} games")
```
